```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregue os datasets
hubs = pd.read_excel('../data/hubs.xlsx')
stores = pd.read_excel('../data/stores.xlsx')
channels = pd.read_csv('../data/channels.csv')
deliveries = pd.read_csv('../data/deliveries.csv')
drivers = pd.read_csv('../data/drivers.csv')
orders = pd.read_csv('../data/orders.csv')
payments = pd.read_csv('../data/payments.csv')

# Crie uma função para plotar boxplots
def plot_boxplots(df, title):
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df.select_dtypes(include=['float64', 'int64']))
    plt.title(title)
    plt.xticks(rotation=45)
    plt.show()

# Plote boxplots para cada DataFrame
plot_boxplots(hubs, 'Boxplot - Hubs')
plot_boxplots(stores, 'Boxplot - Stores')
plot_boxplots(channels, 'Boxplot - Channels')
plot_boxplots(deliveries, 'Boxplot - Deliveries')
plot_boxplots(drivers, 'Boxplot - Drivers')
plot_boxplots(orders, 'Boxplot - Orders')
plot_boxplots(payments, 'Boxplot - Payments')
```
